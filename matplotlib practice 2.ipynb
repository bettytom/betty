```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])# main axes
axes2=fig.add_axes([0.55,0.55,0.3,0.3])
y=np.sin(x)
axes1.plot(x,y,'b')
axes2.plot(x,np.cos(x),'r')
axes2.set_title("cosine")
axes1.set_title("sine")
plt.show()
```

```python
a1=plt.subplot2grid((3,3),(0,0), colspan=2)
a2=plt.subplot2grid((3,3),(0,2), rowspan=3)
a3=plt.subplot2grid((3,3),(1,0),rowspan=2, colspan=2)
x=np.arange(1,10)
a2.plot(x,x*x)
a2.set_title("square")
a1.plot(x,np.exp(x))
a1.set_title("exp")
a3.plot(x,np.log(x))
a3.set_title("log")
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(12,4))
x=np.arange(1,11)
axes[0].plot(x,x**3,'g',linewidth=2)
axes[0].grid()
axes[0].set_title("default grid")

axes[1].plot(x,np.exp(x), 'r')
axes[1].grid('b-',linewidth=0.25)
axes[1].set_title("custom grid")

axes[2].plot(x,x)
axes[2].set_title("no grid")
fig.tight_layout()
plt.show()
```

```python
# formatting axes
fig, axes=plt.subplots(1,2,figsize=(10,4))
x=np.arange(1,5)
axes[0].plot(x,np.exp(x))
axes[0].plot(x,x**2)
axes[0].set_title("normal scale")

axes[1].plot(x,np.exp(x))
axes[1].plot(x,x**2)
axes[1].set_yscale("log") # setting the yscale
axes[1].set_title("Logarithmic scale(y)")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")
axes[0].xaxis.labelpad=10

axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
# creating a dual axes using 'twinx()' and twiny() functions
fig=plt.figure()
a1=fig.add_axes([0,0,1,1])
x=np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel("exp")
a2=a1.twinx()# using twinx()
a2.plot(x,np.log(x),'ro-')
a2.set_ylabel("log")
fig.legend(labels=('exp','log'),loc='upper left')
plt.show()
```

```python
# creating a contour plot
# used to show a 3-D plot on a 2-D plot
import matplotlib.pyplot as plt
import numpy as np
xlist=np.linspace(-3,3,100)
ylist=np.linspace(-3,3,100)
X,Y=np.meshgrid(xlist,ylist)# you have to make this first, it contains both x and y variables
Z=np.sqrt(X**2+Y**2)
fig,ax=plt.subplots(1,1)
cp=ax.contourf(X,Y,Z)# the filled contour plot
fig.colorbar(cp) # add a colorbar to the plot
ax.set_title("Filled Contour Plot")
ax.set_xlabel("x(cm)")
ax.set_ylabel("y(cm)")
plt.show()
```

```python
# creating a quiver plot
x,y=np.meshgrid(np.arange(-2,2,.2), np.arange(-2,2,.25))
z=x*np.exp(-x**2-y**2)
v,u=np.gradient(z,.2,.2)
fig,ax=plt.subplots()
q=ax.quiver(x,y,u,v)
plt.show()
```

```python
# creating a 3Dcontour plot
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from mpl_toolkits import mplot3d
# creating a function first
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour');
```

```python
# creating a 3Dwireframe plot
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='cyan')
ax.set_title('wireframe');
```

```python
# creating a surface plot
x=np.outer(np.linspace(-2,2,30),np.ones(30))
y=x.copy().T # transpose
z=np.cos(x**2+y**2)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='viridis',edgecolor='none')
ax.set_title('Surface plot');
```

```python
# working with text in matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.text(3,8,'boxed italics text in data coords', style='italic', bbox={'facecolor':'cyan'})
ax.text(2,6,r'an equation: $E=mc^2$', fontsize=15,color='m')
ax.text(4,0.05,'colored text in axes coords', verticalalignment='bottom', color='green',
        fontsize=15)
ax.plot([2],[1],'o')
ax.annotate('annotate',xy=(2,1),xytext=(3,4),arrowprops=dict(facecolor='black',shrink=0.05))
ax.axis([0,10,0,10]);
```
