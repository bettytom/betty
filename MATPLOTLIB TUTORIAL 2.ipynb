Of all these functions, ax.fill_between(...) is probably one of the most often. In its most basic form, it fills between the given y-values and 0.

```python
np.random.seed(1)
y=np.random.randn(100).cumsum()
x=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.fill_between(x,y,color='orange')
plt.show()
```

However, it can also be used to fill between two curves. Particularly useful when you want to show an envelope of some sort(e.g, error, amplitude, confidence,etc).

```python
x=np.linspace(0,10,200)
y1=2*x+1
y2=3*x+1.2
y_mean=0.5*x*np.cos(2*x)+2.5*x+1.1
fig,ax=plt.subplots()
#plot the envelope with fill_between
ax.fill_between(x,y1,y2, color='yellow')
#plot the centerline with'plot'
ax.plot(x,y_mean,color='black')
plt.show()
```

## The Data keyword argument
this feature uses the datatype: dictionaries

```python
x= np.linspace(0,10,200)
data_obj={'x':x,
          'y1':2*x+1,
          'y2':3*x+1.2,
          'mean':0.5*x*np.cos(2*x)+2.5*x+1.1}
fig,ax=plt.subplots()
ax.fill_between('x','y1','y2', color='yellow', data=data_obj)
ax.plot('x', 'mean',color='black',data=data_obj)
plt.show()
```

## Shared parameters for imshow,pcolormesh,contour,scatter,etc

1. cmap- the colormap or basically gives the colour to the data input

2.vmin-the minimum data value that will correspond to the 'bottom' of the colormap

3. vmax-the maximum data that will correspond to the 'top' of the colormap.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(1)
# generate random data with different ranges
data1=np.random.random((10,10))
data2=2*np.random.random((10,10))
data3=3*np.random.random((10,10))
# set up your figure and axes
fig,axes=plt.subplots(ncols=3, figsize=plt.figaspect(0.5))
fig.tight_layout() # makes the subplots fill up the figures a bit more
cax=fig.add_axes([0.25,0.1,0.55,0.03]) # the axes for the colorbar
# using a for loop
for ax,data in zip(axes,[data1,data2,data3]):
    im=ax.imshow(data, vmin=0,vmax=3, interpolation='nearest')
fig.colorbar(im, cax=cax, orientation='horizontal')
plt.show()

```
